# Distribution of each component of CVSS base score

```python
labels = ['CVSS3_AtkComp', 'CVSS3_Privilege', 'CVSS3_Interact', 'CVSS3_Scope',
       'CVSS3_Conf', 'CVSS3_Integrity', 'CVSS3_Avail', 'CVSS3_Severity',
       'CVSS2_AccessVect', 'CVSS2_AccessComp', 'CVSS2_Auth',
       'CVSS2_Conf', 'CVSS2_Integrity', 'CVSS2_Avail', 'CVSS2_Severity']
plt.figure(figsize=(20, 30))
plt.subplots_adjust(wspace=0.5, hspace=0.2)

for index in range(len(labels)):
    plt.subplot(6, 3, index + 1)
    sns.countplot(x=labels[index], data=df)
```

# Histograms of base scores in CVSS versions 2 and 3

```python
cvss3 = df[df.CVSS3_Conf.notnull()]
cvss2 = df[df.CVSS2_Conf.notnull()]
plt.figure(figsize=(15, 5))

# CVSS3
ax1 = plt.subplot(1, 2, 1)
ax1.hist(cvss3.CVSS3_BScore, bins=10)
ax1.set_title("Histogram of CVSS3 Base Score (Severity)")

# CVSS2
ax2 = plt.subplot(1, 2, 2)
ax2.hist(cvss2.CVSS2_BScore, bins=10)
ax2.set_title("Histogram of CVSS2 Base Score (Severity)")
```
